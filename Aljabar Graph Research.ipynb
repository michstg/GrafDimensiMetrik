```python
import networkx as nx
V = ['a','b','c','d','e','f','g','h','i','j','k','l'] # simpul
E = [('a','b',23),('b','d',6),('b','g',15),('b','f',7),('c','d',24),('d','i',9),
         ('d','f',11),('e','f',27),('f','k',8),('g','i',12),('g','k',13),('g','h',14),
         ('i','j',25),('i','k',10),('k','l',26)] # sisi & bobot-nya


# Mulai Membentuk Graf-nya
G = nx.Graph() # buat graf kosong


for vertex in V: # Menambahkan semua simpulnya
  G.add_node(vertex)
for v1,v2,w in E: # Menambahkan sisinya
  G.add_edge(v1,v2, weight=w)


print('G: {0} simpul, {1} sisi'.format(G.number_of_nodes(),G.number_of_edges()))
```

```python
pos = nx.spring_layout(G)
eL = nx.get_edge_attributes(G,'bobot')


nx.draw_networkx_nodes(G,pos) 
nx.draw_networkx_labels(G,pos)
nx.draw_networkx_edges(G,pos)
```
